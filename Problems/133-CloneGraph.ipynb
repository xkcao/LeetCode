```python
import collections
# Definition for a Node.
class Node(object):
    def __init__(self, val = 0, neighbors = None):
        self.val = val
        self.neighbors = neighbors if neighbors is not None else []


class Solution(object):
    def cloneGraph(self, node):
        """
        :type node: Node
        :rtype: Node
        """
        #base case, node=None
        if node==None:
            return None
        
        #use a hashMap to map value to the new node
        valueToNewNode={}
        #use a set to mark the visited nodes
        visitedNodes=set()
        
        #use a queue, FIFO
        dq = collections.deque()
        #push the first node into the queue
        dq.append(node)
        
        newFirstNode=None
        
        #while the queue is not empty
        while len(dq)!=0:
            #pop the queque, get the currentNode
            currentNode=dq.popleft()
            #fetcth or create a newNode with the value the same as currentNode
            if currentNode.val in valueToNewNode:
                newNode = valueToNewNode[currentNode.val]
            else:
                newNode = Node(currentNode.val)
                # Hashmap[value]=newNode
                valueToNewNode[newNode.val]=newNode
            
            if newFirstNode==None:
                newFirstNode=newNode
            
            #for each neighbourNode in neighbors of currentNode
            for neighborNode in currentNode.neighbors:
                if neighborNode not in visitedNodes:
                    dq.append(neighborNode)
                    visitedNodes.add(neighborNode)
                                    
                if neighborNode.val in valueToNewNode:
                    newNeighbourNode = valueToNewNode[neighborNode.val]
                else:
                    newNeighbourNode = Node(neighborNode.val)
                    valueToNewNode[newNeighbourNode.val]=newNeighbourNode

                newNode.neighbors.append(newNeighbourNode)                
                    
            visitedNodes.add(currentNode)
                #if the neighbourNode has already been fully visited
                    #continue
                #else 
                    #push the neighbourNode into he queue
                    #fetch from Hashmap or create a newNeighbourNode with the same value as neighbourNode
                    #append newNeighbourNode into the neightbors of newNode
                    #Hashmap[value]=newNeighbourNode
            
            #mark currentNode as visited
            
        #return the reference to the new graph
        return newFirstNode
```
