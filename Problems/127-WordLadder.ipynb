```python
import heapq
import sys
class Solution(object):
    def ladderLength(self, beginWord, endWord, wordList):
        """
        :type beginWord: str
        :type endWord: str
        :type wordList: List[str]
        :rtype: int
        """
        #return true if two words differ by only 1 letter
        def differOneLetter(word1,word2):
            if len(word1)!=len(word2):
                return False
            
            numDiff=0
            for i in range(len(word1)):
                if word1[i]!=word2[i]:
                    numDiff+=1
                    
                    if numDiff>1:
                        return False
            
            return numDiff==1
        
        #special case
        if endWord not in wordList:
            return 0
        
        #for each word, store a list of words which it differs only by 1 letter
        dictAdjacent={}
        for word1 in wordList:
            dictAdjacent[word1]=[]
            for word2 in wordList:
                if differOneLetter(word1,word2):
                    dictAdjacent[word1].append(word2)
        
        dictAdjacent[beginWord]=[]
        for word in wordList:
            if differOneLetter(beginWord,word):
                dictAdjacent[beginWord].append(word)
        
        #Dijkstra        
        shortestPath={}
        for word in wordList:
            shortestPath[word]=sys.maxsize
        shortestPath[beginWord]=1
        #a heap for remaining words with key as current shortest distance        
        toVisit=[]
        visited={}
        heapq.heappush(toVisit,(shortestPath[beginWord],beginWord))    

        while len(toVisit)!=0:
            item=heapq.heappop(toVisit)
            word=item[1]
            if word in visited:
                continue
            visited[word]=1

            neighbors=dictAdjacent[word]
            for neighbor in neighbors:
                if neighbor in visited:
                    continue
                if shortestPath[word]+1<shortestPath[neighbor]:                    
                    shortestPath[neighbor]=shortestPath[word]+1
                    heapq.heappush(toVisit,(shortestPath[neighbor],neighbor))
                
        return shortestPath[endWord] if shortestPath[endWord]!=sys.maxsize else 0
```

```python
from collections import defaultdict
class Solution(object):
    def ladderLength(self, beginWord, endWord, wordList):
        """
        :type beginWord: str
        :type endWord: str
        :type wordList: List[str]
        :rtype: int
        """

        if endWord not in wordList or not endWord or not beginWord or not wordList:
            return 0

        # Since all words are of same length.
        L = len(beginWord)

        # Dictionary to hold combination of words that can be formed,
        # from any given word. By changing one letter at a time.
        all_combo_dict = defaultdict(list)
        for word in wordList:
            for i in range(L):
                # Key is the generic word
                # Value is a list of words which have the same intermediate generic word.
                all_combo_dict[word[:i] + "*" + word[i+1:]].append(word)


        # Queue for BFS
        queue = collections.deque([(beginWord, 1)])
        # Visited to make sure we don't repeat processing same word.
        visited = {beginWord: True}
        while queue:
            current_word, level = queue.popleft()
            for i in range(L):
                # Intermediate words for current word
                intermediate_word = current_word[:i] + "*" + current_word[i+1:]

                # Next states are all the words which share the same intermediate state.
                for word in all_combo_dict[intermediate_word]:
                    # If at any point if we find what we are looking for
                    # i.e. the end word - we can return with the answer.
                    if word == endWord:
                        return level + 1
                    # Otherwise, add it to the BFS Queue. Also mark it visited
                    if word not in visited:
                        visited[word] = True
                        queue.append((word, level + 1))
                all_combo_dict[intermediate_word] = []
        return 0
```
