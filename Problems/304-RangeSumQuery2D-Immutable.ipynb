```python
class NumMatrix(object):

    def __init__(self, matrix):
        """
        :type matrix: List[List[int]]
        """
        #calcualte the sum of region from origin (Top left corner) to each cell in the matrix
        # and store them in a hash table
        self.sumRegionFromOrigin={}
        
        m=len(matrix)
        n=len(matrix[0])
        
        for i in range(m):
            sumRow=0
            for j in range(n):
                sumRow+=matrix[i][j]
                if i==0:
                    self.sumRegionFromOrigin[(i,j)]=sumRow
                else:
                    self.sumRegionFromOrigin[(i,j)]=self.sumRegionFromOrigin[(i-1,j)]+sumRow
        

    def sumRegion(self, row1, col1, row2, col2):
        """
        :type row1: int
        :type col1: int
        :type row2: int
        :type col2: int
        :rtype: int
        """
        res= self.sumRegionFromOrigin[(row2,col2)]
        
        #if row1 is not first row
        if row1>=1:
            res-=self.sumRegionFromOrigin[(row1-1,col2)]
            
        #if col1 is not first column
        if col1>=1:
            res-=self.sumRegionFromOrigin[(row2,col1-1)]
            
        #if row1 is not first row and col1 is not first column
        if row1>=1 and col1>=1:
            res+=self.sumRegionFromOrigin[(row1-1,col1-1)]
                
        return res
        


# Your NumMatrix object will be instantiated and called as such:
# obj = NumMatrix(matrix)
# param_1 = obj.sumRegion(row1,col1,row2,col2)
```
