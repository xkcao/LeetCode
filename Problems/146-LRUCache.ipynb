```python
class LRUCache(object):
    capacity=0
    mapping={}
    recentUse=[]

    def __init__(self, capacity):
        """
        :type capacity: int
        """
        self.capacity=capacity
        self.mapping={}
        self.recentUse = []
        

    def get(self, key):
        """
        :type key: int
        :rtype: int
        """
        if key in self.mapping:
            self.recentUse.remove(key)
            self.recentUse.append(key)
            return self.mapping[key]        
        else:
            return -1
        

    def put(self, key, value):
        """
        :type key: int
        :type value: int
        :rtype: None
        """
        
        if key in self.mapping:
            self.mapping[key]=value
            self.recentUse.remove(key)
            self.recentUse.append(key)            
        elif len(self.mapping)<self.capacity:
            self.mapping[key]=value
            self.recentUse.append(key)
        else:
            toRemove=self.recentUse.pop(0)
            self.recentUse.append(key)
            self.mapping.pop(toRemove)
            self.mapping[key]=value
            
            


# Your LRUCache object will be instantiated and called as such:
# obj = LRUCache(capacity)
# param_1 = obj.get(key)
# obj.put(key,value)
```
